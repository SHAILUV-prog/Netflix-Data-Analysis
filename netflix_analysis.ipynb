```python
import seaborn as sns
```

```python
df=pd.read_csv("netflix_titles.csv")
```

```python
df['date_added']=pd.to_datetime(df['date_added'],errors='coerce')
```

```python
df['year_added']=df['date_added'].dt.year
```

```python
df['month_added']=df['date_added'].dt.month
```

```python
df['country'].fillna('Unknown',inplace=True)
```

```python
df['rating'].fillna('Unknown',inplace=True)
```

```python
a=sns.countplot(x='type',data=df)
a.set_title("Count of Movies and TV Shows on Netflix",fontsize=16)
plt.show()
```

```python
df['country'].value_counts().head(5).plot(kind='bar')
plt.title("Top 5 counries producing content for Netflix",fontsize=16)
plt.show()
```

```python
df['year_added'].value_counts().sort_index().plot(kind='line')
plt.title("Content added on netflix over the years")
plt.show()
```

```python
df['listed_in'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 generes on netflix")
plt.show()
```

```python
df['rating'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 ratings on netflix")
plt.show()
```
